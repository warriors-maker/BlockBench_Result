```python
import pandas as pd
import numpy as np
import re 
import csv
import sklearn.externals.joblib as joblib
from io import StringIO
from sklearn.externals.joblib import Memory
memory = Memory(cachedir='/tmp', verbose=0)
```

```python
def return_digit(words):
    count = 0;
    for c in words:
        if not c.isdigit():
            break;
        count += 1;
    return words[:count].strip()

@memory.cache
def convertToCSV(filenames, numNodes, csv=True):
    # print header
    csvOutput = []
    if csv:
        csvOutput.append("nodes,pollIntervall,txCount,latency,avgLatency")
    for filename in filenames:
        #nodes = re.search('nodes(\d+)_', filename)
        #numNodes = nodes.group(1) if nodes else 2
        with open(filename) as f:
            for l in f:
#                 print(l)
                if "tx count" in l:
                    my_list = l.split()
                    for index, words in enumerate(my_list):
                        if words == 'count':
                            if index + 5 >= len(my_list):
                                continue
                            throughput = return_digit(my_list[index + 2])
                            latency = return_digit(my_list[index + 5])
                            if re.search('[a-zA-Z]', throughput) or re.search('[a-zA-Z]', latency) or len(throughput) == 0 or len(latency) == 0:
                                continue
                            else:
                                avgLatency = (float(latency) / float(throughput)) if float(throughput) != 0 else 0
                                line = [str(3), str(2), str(throughput), str(latency), str(avgLatency)]
#                                 print(line)
                                if csv:
                                    csvOutput.append(','.join(line))
                                else:
                                    csvOutput.append([ float(x) for x in line ])
    return csvOutput
```

# Normal_25 Threads 100 transaction / second

```python
pwd = "./0_hl_exp_3_servers_25_threads_100_rates"
csvOut_overall = convertToCSV([pwd + '/client_10.150.0.26_25',
                       pwd + '/client_10.150.0.29_25',
                       pwd +'/client_10.150.0.30_25',
                       ], 3, False)
df_overall = pd.DataFrame.from_records(csvOut_overall, columns=['nodes','pollIntervall','txCount','latency','avgLatency'])
df_overall.agg(['min', 'max', 'mean','sum'])
```
